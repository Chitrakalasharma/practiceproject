```python
x=df.iloc[:,0:-1]
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=55)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
df.columns
```

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print("predicted result salary:",pred)
print("actual salary",y_test)
```

```python
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))   
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root MeanSquared Error:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
df1=pd.DataFrame(pred)
df1.to_csv("Salary_pred.csv")
```
